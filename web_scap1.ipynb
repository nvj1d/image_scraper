```python
from bs4 import BeautifulSoup
import requests
import os
```

```python
url = 'https://www.autos-motos.net/'
damaged_url = 'https://www.autos-motos.net/damaged-cars?page='
```

```python
def get_car(full_url, path):
  r = requests.get(full_url)
  soup = BeautifulSoup(r.text, 'html.parser')
  images = soup.find_all('img', alt=True)
  os.chdir('/content/drive/MyDrive/scrap/')
  dir = soup.title.text
  dir = dir.replace('/','').replace(' ','').replace('.-WWW.AUTOS-MOTOS.NETen','')
  if os.path.isdir(dir) is False:
    os.mkdir(dir)
  print(dir + ' created ..')
  os.chdir(dir)
  # print(os.getcwd())
  print('adding images to folder ..')
  for image in images:
    word = '/pictures/'
    name = image['alt']
    link = full_url + image['src']
    link = link.replace(path,'').replace('-S~','-B~')
    if word in link:
      name = name.replace(' ','').replace('/','')
      print(name + '.jpg')
      img = open(name + '.jpg', 'wb')
      with img as f:
        im = requests.get(link)
        f.write(im.content)
  print('adding images to ' + dir + ' finished!')
```

```python
num_pages = 62
for i in range(16, num_pages + 1):
  page_url = damaged_url + str(i)
  print(page_url)
  re = requests.get(page_url)
  soup = BeautifulSoup(re.text, 'html.parser')
  for link in soup.find_all('a', href=True):
    if '/car/' in link.get('href'):
      path = link.get('href').split("/")[2]
      print(path) #works fine!
      get_car(url+path, path)
```
